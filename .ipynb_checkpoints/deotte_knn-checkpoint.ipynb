```python
import numpy as np, pandas as pd
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt
```

```python
path_to_data = 'e:\\Krivenko\\Kaggle\\2021\\New20211005\\'
train_name = 'train.csv'
test_name = 'test.csv'
train = pd.read_csv(path_to_data + train_name)
test = pd.read_csv(path_to_data + test_name)
```

```python
exhale = 80-train.groupby('breath_id')[['u_out']].agg('sum')
length = train.groupby('breath_id')[['time_step']].agg('max')
print('Train shape:',train.shape)
train.head()
```

```python
# CONVERT TRAIN TO SERIES
series = train.groupby('breath_id').agg(list).reset_index()
for k in range(80): series[f'x_{k}'] = series.u_in.apply(lambda x: x[k])
for k in range(80): series[f'y_{k}'] = series.pressure.apply(lambda x: x[k])
for k in range(80): series[f'z_{k}'] = 1-series.u_out.apply(lambda x: x[k])
series.R = series.R.apply(lambda x: x[0])
series.C = series.C.apply(lambda x: x[0])
series = series.drop(['id','time_step','u_in','u_out','pressure'],axis=1)
series = series.merge(exhale,on='breath_id',how='left')
series = series.merge(length,on='breath_id',how='left')
series = series.rename({'time_step':'time_length','u_out':'expire'},axis=1)
series = series.sort_values('breath_id').reset_index(drop=True)

print('Train as series shape:', series.shape )
print('Min inhale length=', series['expire'].min(),',Max inhale length=', series['expire'].max(),
      'Max breath length=',series['time_length'].max() )
series.head()
```

```python
# FIRST 3 COLUMNS OF TRAIN SERIES DATAFRAME NEED TO BE IGNORED
IGNORE = 3
# MORE NEIGHBORS THAN WE NEED FOR ALL DISPLAYS
NEIGHBORS = 500
# HOW MANY TIME_STEPS TO USE FOR KNN COMPARISON
TIME_STEPS = 80
# HOW MANY INITIAL TIME_STEPS TO SKIP FOR KNN COMPARISON
SKIP = 0 
```

```python
model = NearestNeighbors(n_neighbors=NEIGHBORS, metric='l1')
model.fit(series.iloc[:,IGNORE+SKIP:IGNORE+TIME_STEPS+SKIP])
```

```python
SHOW = [24,68,75]
CTS = [300]*len(SHOW)
```

```python
for USE,CT in zip(SHOW,CTS):
    distances, indices = model.kneighbors(series.iloc[USE:USE+1,IGNORE+SKIP:IGNORE+TIME_STEPS+SKIP])
        
    plt.figure(figsize=(20,4))
    plt.plot(np.arange(80),series.iloc[USE,IGNORE:IGNORE+80].values,label='u_in')
    plt.plot(np.arange(80),series.iloc[USE,IGNORE+80:IGNORE+160].values,label='pressure')
    y_max = plt.ylim()[1]
    if TIME_STEPS!=80: plt.plot([SKIP+TIME_STEPS-1,SKIP+TIME_STEPS-1],[0,y_max],'--',color='gray')
    if SKIP!=0: plt.plot([SKIP-1,SKIP-1],[0,y_max],'--',color='gray')
    exhale = series.loc[USE,'expire']
    plt.plot([exhale,exhale],[0,y_max],'--',color='black',label='exhale')
    rr = series.loc[USE,'R']; cc = series.loc[USE,'C']; 
    bb = series.loc[USE,'breath_id']; tt = series.loc[USE,'time_length']
    plt.title(f'Breath_Id={bb}, R={rr}, C={cc}, Length={tt:.3}',size=16)
    plt.legend()
    
    temp2 = series.iloc[indices[0]].reset_index(drop=True)
    cdict = {5: 'yellow', 20: 'orange', 50: 'red'}
    plt.figure(figsize=(20,10))
    legend = {5:0, 20:0, 50:0}
    for r in range(CT):
        if r==0: plt.plot(np.arange(80), temp2.iloc[r,IGNORE:IGNORE+80].values, color='blue', label='u_in')
        else: plt.plot(np.arange(80), temp2.iloc[r,IGNORE:IGNORE+80].values, color='blue')
        if legend[temp2.loc[r,'R']]==0:
            legend[temp2.loc[r,'R']]=1; cc = temp2.loc[r,'R']
            plt.plot(np.arange(80), temp2.iloc[r,IGNORE+80:IGNORE+160].values, 
                     c=cdict[temp2.loc[r,'R']], label=f'R={cc}')
        else: plt.plot(np.arange(80), temp2.iloc[r,IGNORE+80:IGNORE+160].values, c=cdict[temp2.loc[r,'R']])
    y_max = plt.ylim()[1]
    if TIME_STEPS!=80: plt.plot([SKIP+TIME_STEPS-1,SKIP+TIME_STEPS-1],[0,y_max],'--',color='gray')
    if SKIP!=0: plt.plot([SKIP-1,SKIP-1],[0,y_max],'--',color='gray')
    plt.title(f'{CT} time series similar to breath_Id={bb}. For each new time series we plot and color code its pressure with respect to R variable',size=16)
    plt.legend()
    plt.show()
    
    temp2 = series.iloc[indices[0]].reset_index(drop=True)
    cdict = {10: 'red', 20: 'orange', 50: 'yellow'}
    plt.figure(figsize=(20,10))
    legend = {10:0, 20:0, 50:0}
    for r in range(CT):
        if r==0: plt.plot(np.arange(80), temp2.iloc[r,IGNORE:IGNORE+80].values, color='blue',label='u_in')
        else: plt.plot(np.arange(80), temp2.iloc[r,IGNORE:IGNORE+80].values, color='blue')
        if legend[temp2.loc[r,'C']]==0:
            legend[temp2.loc[r,'C']]=1; cc = temp2.loc[r,'C']
            plt.plot(np.arange(80), temp2.iloc[r,IGNORE+80:IGNORE+160].values, 
                     c=cdict[temp2.loc[r,'C']], label=f'C={cc}')
        else: plt.plot(np.arange(80), temp2.iloc[r,IGNORE+80:IGNORE+160].values, c=cdict[temp2.loc[r,'C']])
    y_max = plt.ylim()[1]
    if TIME_STEPS!=80: plt.plot([SKIP+TIME_STEPS-1,SKIP+TIME_STEPS-1],[0,y_max],'--',color='gray')
    if SKIP!=0: plt.plot([SKIP-1,SKIP-1],[0,y_max],'--',color='gray')
    plt.title(f'{CT} time series similar to breath_Id={bb}. For each new time series we plot and color code its pressure with respect to C variable',size=16)
    plt.legend()
    plt.show()
```

```python
# LOAD TEST
test = pd.read_csv(path_to_data + 'test.csv')
exhale = 80-test.groupby('breath_id')[['u_out']].agg('sum')
length = test.groupby('breath_id')[['time_step']].agg('max')
first_id = test.groupby('breath_id')[['id']].agg('min')
print('Test shape:',test.shape)
test.head()
```

```python
# CONVERT TEST TO SERIES
test_series = test.groupby('breath_id').agg(list).reset_index()
for k in range(80): test_series[f'x_{k}'] = test_series.u_in.apply(lambda x: x[k])
for k in range(80): test_series[f'z_{k}'] = 1-test_series.u_out.apply(lambda x: x[k])
test_series.R = test_series.R.apply(lambda x: x[0])
test_series.C = test_series.C.apply(lambda x: x[0])
test_series = test_series.drop(['id','time_step','u_in','u_out'],axis=1)
test_series = test_series.merge(exhale,on='breath_id',how='left')
test_series = test_series.merge(length,on='breath_id',how='left')
test_series = test_series.merge(first_id,on='breath_id',how='left')
test_series = test_series.sort_values('breath_id').reset_index(drop=True)
test_series = test_series.rename({'time_step':'time_length','u_out':'expire','id':'first_id'},axis=1)

print('Test as series shape:', test_series.shape )
print('Min inhale length=', test_series['expire'].min(),',Max inhale length=', test_series['expire'].max(),
      'Max breath length=',test_series['time_length'].max() )
test_series.head()
```

```python
NEIGHBORS = 5
SIZE = 80

# OOF
all_true = []
all_oof = []
all_true_inhale = []
all_oof_inhale = []

# TEST PREDS
all_pred = []
all_test_id = []

for r in [5,20,50]:
    for c in [10,20,50]:
        
        # FIT TRAIN KNN
        temp2 = series.loc[(series.R==r)&(series.C==c)].reset_index(drop=True)
        model = NearestNeighbors(n_neighbors=NEIGHBORS, metric='l1')
        model.fit(temp2.iloc[:,IGNORE:SIZE+IGNORE])

        # INFER OOF
        err = []
        distances, indices = model.kneighbors(temp2.iloc[:,IGNORE:IGNORE+SIZE])
        
        # INFER TEST
        test_temp2 = test_series.loc[(test_series.R==r)&(test_series.C==c)].reset_index(drop=True)
        test_distances, test_indices = model.kneighbors(test_temp2.iloc[:,IGNORE:IGNORE+SIZE])
        
        for DO in range(80):
            
            # OOF AND COLLECT TRUE (BELOW WE USE COLUMN 1 NOT 0 TO AVOID USING ITSELF)
            oof = temp2.loc[indices[:,1],f'y_{DO}'].values                        
            true = temp2[f'y_{DO}'].values
            mask = temp2[f'z_{DO}'].values
            
            # TEST PREDS (BELOW WE USE COLUMN 0 NOT 1, BECAUSE THERE IS NO ITSELF)
            preds = temp2.loc[test_indices[:,0],f'y_{DO}'].values
            test_ids = test_temp2.first_id.values+DO

            # RECORD OOF ERROR FOR PLOT
            mae = np.mean(np.abs( true-oof ))
            err.append(mae)
            
            # RECORD OOF
            idx = np.where( mask==1 )[0]
            all_true_inhale.append( true[idx] )
            all_oof_inhale.append( oof[idx] )
            all_true.append( true )
            all_oof.append( oof )
            
            # RECORD TEST PREDS
            all_pred.append( preds )
            all_test_id.append( test_ids )
            
        # PLOT OOF ERROR
        plt.figure(figsize=(20,5))
        plt.plot(np.arange(80),err,'-o')
        plt.ylabel('OOF RSME',size=14)
        plt.xlabel('Time Step',size=14)
        plt.title(f'OOF RSME for R={r} C={c}',size=16)
        plt.show()
```

```python
def check_optimal_neighbors(namount=50):
    total_true_inhale = []
    total_oof_inhale = []
    total_true = []
    total_oof =[]
    
    for ii in range(namount-2):
        # My median
        NEIGHBORS = ii+2
        SIZE = 80

        # OOF
        all_true = []
        all_oof = []
        all_true_inhale = []
        all_oof_inhale = []

        # TEST PREDS
        all_pred = []
        all_test_id = []

        for r in [5,20,50]:
            for c in [10,20,50]:
        
                # FIT TRAIN KNN
                temp2 = series.loc[(series.R==r)&(series.C==c)].reset_index(drop=True)
                model = NearestNeighbors(n_neighbors=NEIGHBORS, metric='l1')
                model.fit(temp2.iloc[:,IGNORE:SIZE+IGNORE])

                # INFER OOF
                err = []
                distances, indices = model.kneighbors(temp2.iloc[:,IGNORE:IGNORE+SIZE])
        
                # INFER TEST
                test_temp2 = test_series.loc[(test_series.R==r)&(test_series.C==c)].reset_index(drop=True)
                test_distances, test_indices = model.kneighbors(test_temp2.iloc[:,IGNORE:IGNORE+SIZE])
        
                for DO in range(80):
            
                    # OOF AND COLLECT TRUE (BELOW WE USE COLUMN 1 NOT 0 TO AVOID USING ITSELF)
                    #oof = temp2.loc[indices[:,1],f'y_{DO}'].values                        
                    oof= []
                    for i in range(indices.shape[1] - 1):
                        oof_local = temp2.loc[indices[:, i+1],f'y_{DO}'].values
                        oof.append(oof_local)
                    oof = np.array(oof).T
                    oof = np.median(oof, axis=-1)
                    
                    true = temp2[f'y_{DO}'].values
                    mask = temp2[f'z_{DO}'].values
            
                    # TEST PREDS (BELOW WE USE COLUMN 0 NOT 1, BECAUSE THERE IS NO ITSELF)
                    #preds = temp2.loc[test_indices[:,0],f'y_{DO}'].values
                    preds= []
                    for i in range(test_indices.shape[1]):
                        preds_local = temp2.loc[test_indices[:, i],f'y_{DO}'].values
                        preds.append(preds_local)
                    preds = np.array(preds).T
                    preds = np.median(preds, axis=-1)
                    test_ids = test_temp2.first_id.values+DO

                    # RECORD OOF ERROR FOR PLOT
                    mae = np.mean(np.abs( true-oof ))
                    err.append(mae)
                    
                    # RECORD OOF
                    idx = np.where( mask==1 )[0]
                    all_true_inhale.append( true[idx] )
                    all_oof_inhale.append( oof[idx] )
                    all_true.append( true )
                    all_oof.append( oof )
                    
                    # RECORD TEST PREDS
                    all_pred.append( preds )
                    all_test_id.append( test_ids )
                    
        total_true_inhale.append(all_true_inhale)
        total_oof_inhale.append(all_oof_inhale)
        total_true.append(all_true)
        total_oof.append(all_oof)
    return total_true_inhale, total_oof_inhale, total_true, total_oof
```

```python
nmax = 20
total_true_inhale, total_oof_inhale, total_true, total_oof = check_optimal_neighbors(nmax)
```

```python
mae_inhale = []
mae_all = []
for tti, toi, tt, to in zip(total_true_inhale, total_oof_inhale, total_true, total_oof):
    # COMPUTE OOF SCORE FOR INHALE ONLY
    all_true2 = np.concatenate(tti)
    all_oof2 = np.concatenate(toi)
    rsme = np.mean(np.abs( all_true2-all_oof2 ))
    print('RAPIDS KNN CV MAE=',rsme, '(inhale only)')
    mae_inhale.append(rsme)

    # COMPUTE OOF SCORE FOR ALL 80 TIME STEPS
    all_true2 = np.concatenate(tt)
    all_oof2 = np.concatenate(to)
    rsme = np.mean(np.abs( all_true2-all_oof2 ))
    print('RAPIDS KNN CV MAE=',rsme, '(all breath)')
    mae_all.append(rsme)
    
plt.figure(figsize=(20,5))
plt.plot(np.arange(nmax-2)+2,np.array(mae_inhale),'-o', label='INHALE')
plt.plot(np.arange(nmax-2)+2,np.array(mae_all),'--x',label='ALL')
plt.ylabel('MAE',size=14)
plt.xlabel('Time Step',size=14)
plt.legend()
plt.title('neighbors vs mae',size=16)
plt.show()
```
